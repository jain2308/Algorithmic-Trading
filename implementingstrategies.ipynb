```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
ticker = yf.Ticker("AAPL")
data = ticker.history(period="1y")
print(data.head())
```

```python
daily_close = data[['Close']]
daily_return = daily_close.pct_change()
daily_return.fillna(0, inplace=True)
print(daily_return)
```

```python
price = data['Close']
mav = price.rolling(window=50).mean()
print(mav)
```

```python
import matplotlib.pyplot as plt
plt.plot(price)
plt.show()
plt.plot(mav)
plt.show()
```

```python
short_lb = 50
long_lb = 120
```

```python
signal_df = pd.DataFrame(index=data.index)
signal_df['signal'] = 0.0
```

```python
signal_df['short_mav'] = data['Close'].rolling(window=short_lb, min_periods=1, center=False).mean()
signal_df['long_mav'] = data['Close'].rolling(window=long_lb, min_periods=1, center=False).mean()
```

window_size = 20

```python
signal_df['signal'][short_lb:] = np.where(signal_df['short_mav'][short_lb:] > signal_df['long_mav'][short_lb:], 1.0, 0.0)
signal_df['positions'] = signal_df['signal'].diff()
```

```python
fig = plt.figure()

plt1 = fig.add_subplot(111, ylabel='Price in$')
data['Close'].plot(ax=plt1, color='r', lw=2.)

signal_df[['short_mav', 'long_mav']].plot(ax=plt1, lw=2., figsize=(12,8))

plt1.plot(signal_df.loc[signal_df.positions == -1.0].index,

signal_df.short_mav[signal_df.positions == -1.0],

'v', markersize=10, color='k')

plt1.plot(signal_df.loc[signal_df.positions == 1.0].index,

signal_df.short_mav[signal_df.positions == 1.0],

'^', markersize=10, color='m')
plt.show()
```

```python
window_size = 20
```

```python
new_df = pd.DataFrame(index=data.index)
new_df['signal'] = 0.0
new_df['mav'] = data['Close'].rolling(window=window_size, min_periods=1, center=False).mean()
new_df['price'] = data['Close']
new_df
```

```python
new_df['signal'] = (new_df['mav'] - new_df['price'])*100
new_df
```

```python
preffered_diff = 200
new_df['position'] = np.where(new_df['signal'] > preffered_diff, 1.0, 0.0)
new_df['position'] = np.where(new_df['signal'] < -preffered_diff, new_df['position']-1, new_df['position'])
new_df
```

```python
plt.figure(figsize=(14, 7))
plt.plot(new_df['price'], label='Price', color='black')
plt.plot(new_df['mav'], label='mav', linestyle='--', alpha=0.7, color='blue')
buy_signals = new_df[new_df['signal'] > preffered_diff]
plt.plot(buy_signals.index, new_df.loc[buy_signals.index, 'price'], '^', markersize=10, color='green', label='Buy Signal')
sell_signals = new_df[new_df['signal'] < -preffered_diff]
plt.plot(sell_signals.index, new_df.loc[sell_signals.index, 'price'], 'v', markersize=10, color='red', label='Sell Signal')
plt.title('Mean Reversion Strategy: Price, Position, and Signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
